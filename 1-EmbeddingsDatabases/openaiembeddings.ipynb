## Cosine Simillarity with Open AI Embeddings

```python
# Finding similar sentences
sentences=[
    "The cat sat on the mat",
    "The dog chased the cat",
    "The bird flew over the cat",
    "Python is a easy language",
    "Machine Learnig is a demanding topic"
]
```

```python
import numpy as np
def cosine_similarity(vec1,vec2):
    """calculate the cosine similaarity between two vectors by creating our own  function
    """
    dot_product=np.dot(vec1,vec2)
    norm1=np.linalg.norm(vec1)
    norm2=np.linalg.norm(vec2)
    return dot_product/(norm1*norm2)
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings(model="text-embedding-3-small")

embeddings
```

```python
sentence_embeddings=embeddings.embed_documents(sentences)
simillarity=[]
for i in range(len(sentences)):
    for j in range(i+1,len(sentences)):
        simillarity_score=cosine_similarity(sentence_embeddings[i],sentence_embeddings[j])

        print(f"{sentences[i]} vs {sentences[j]}  \n simillarity score: {simillarity_score}\n")
```

```python
# List of sentences (corpus)
documents = [
    "The cat sat on the mat.",
    "The dog chased the cat in the garden.",
    "Birds are flying across the blue sky.",
    "Python is a popular programming language.",
    "Machine learning is transforming the tech industry."
]

# Query sentence
query = "I want to learn about programming languages."
```

```python
def semantic_search(query, documents,embedding_model, top_k=3):
    """Semantic search simple implement"""
    # embed the query
    quer_embeddings=embedding_model.embed_query(query)

    # embed the documents
    document_embeddings=embedding_model.embed_documents(documents)

    simillarity=[]
    for i in range(len(documents)):
        simillarity_score=cosine_similarity(quer_embeddings,document_embeddings[i])
        simillarity.append(( simillarity_score,documents[i],))
    
    #Sort the simillarity
    simillarity.sort(reverse=True)
    return simillarity[:top_k]
```

```python
result=semantic_search(query,documents,embeddings)
result
```
