## Create Embedding Using Huggingface opensource modeel

```python
from langchain_huggingface import HuggingFaceEmbeddings

# Create embeddings

embedings=HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
)
embedings
```

```python
text="Hi,I am learning about embeddings"
embeded_vector=embedings.embed_query(text)
print(f"embeded_vector dimension is {len(embeded_vector)}")
print(f"Vector representation is: {embeded_vector}")
```

```python
sentences=[
    "The cat sat on the mat",
    "The dog played in the yard",
    "you are very beautful"
]
embeded_vectors=embedings.embed_documents(sentences)
print(f"embeded_vector dimension is {len(embeded_vectors)}")
print(f"Vector representation is: {embeded_vectors[0]}")
```
